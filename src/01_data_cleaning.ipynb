### DraftKings
Looking at the DraftKings files, it looks like there's 2 tables so we want to split and save those off as separate tables

````python
dk_file_list = glob.glob("../raw_data/draftkings/*")
````

````python
dk_file_list
````

````python
i = 0
for file in dk_file_list:
    raw_df = pd.read_csv(file)
    
    # filter to scoring df columns (first 6 cols)
    scoring_df = raw_df[['Rank','EntryId','EntryName','TimeRemaining','Points','Lineup']]
    scoring_df['game_id'] = i
    
    # filter to player df columns (last 4 cols)
    # dropping NA cuz it has a lot less rows than the scoring df
    player_df = raw_df[['Player','Roster Position','%Drafted','FPTS']].dropna()
    player_df['game_id'] = i
    
    # union all the datasets together
    if i == 0:
        combined_scoring_df = scoring_df
        combined_player_df = player_df
    else:
        combined_scoring_df = pd.merge(combined_scoring_df, scoring_df, how = 'outer')
        combined_player_df = pd.merge(combined_player_df, player_df, how = 'outer')
    i+=1
    
combined_scoring_df.to_csv("../cleaned_data/roster_df.csv", index = False)
combined_player_df.to_csv("../cleaned_data/player_df.csv", index = False)
````
